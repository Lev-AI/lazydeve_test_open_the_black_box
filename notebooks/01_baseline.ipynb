# Baseline Model Training and Evaluation

In this notebook, we will load the dataset, train a baseline model using RandomForest or XGBoost, and log the results with MLflow.

```python
import mlflow
import mlflow.sklearn
from src.data_loader import load_data
from src.baseline_model import train_baseline

# Load dataset
data = load_data('data/dataset.csv')
X_train, y_train = data['X_train'], data['y_train']

# Train baseline model
model = train_baseline(X_train, y_train, model_type='rf')

# Log model with MLflow
mlflow.start_run()
mlflow.sklearn.log_model(model, 'model')
mlflow.log_param('model_type', 'RandomForest')
mlflow.end_run()
```
