# Data Mutations

In this notebook, we will apply various mutations to the original dataset using the mutation engine.

```python
import pandas as pd
from src.data_loader import load_data
from src.mutation_engine import add_noise, drop_features, drift_features

# Load original dataset
data = load_data('data/dataset.csv')
X = data['X']

# Apply mutations
X_noisy = add_noise(X, intensity=0.05)
X_dropped = drop_features(X_noisy, drop_ratio=0.1)
X_drifted = drift_features(X_dropped, drift_ratio=0.1)

# Visualize mutated data
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.title('Mutated Data Visualization')
plt.scatter(X_drifted[:, 0], X_drifted[:, 1])
plt.show()
```
